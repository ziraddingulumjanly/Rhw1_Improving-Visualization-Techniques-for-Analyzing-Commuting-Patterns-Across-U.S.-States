```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Given csv dataset
file_path = 'Commute.csv'  
df = pd.read_csv(file_path)

# Specific states
states = ['California', 'Colorado', 'District of Columbia',
                    'Florida', 'New York', 'Texas']

# Data filtering for specific states (six in our case)
df_filtered = df[df['State'].isin(states)] 
# note: df_filterd is a subset of states same as those in the pie chart: 
# California, Colorado, District of Columbia, Florida, New York, Texas)

# Pivot the data to have modes as columns for stacked bar plotting
pivot_data = df_filtered.pivot(index='State', columns='Mode', 
                               values='ProportionCommute').fillna(0)

# Data sort by State order
pivot_data = pivot_data.loc[states]

# Display of bargraph
plt.figure(figsize=(12,8))
pivot_data.plot(kind='bar', stacked=True, colormap='tab20', figsize=(12,4))
plt.title('Commute Mode Proportions by State')
plt.xlabel('State')
plt.ylabel('Proportion')
plt.legend(title='Commute Mode', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
df_filtered = df[df['State'].isin(states)]

# Bar chart
plt.figure(figsize=(12, 4))
sns.barplot(x='Mode', y='ProportionCommute', hue='State', data=df_filtered)
plt.title('Proportion of Different Transport Modes Used specific States')
plt.xlabel('Transport Mode')
plt.ylabel('Proportion of Commuters')
plt.xticks(rotation=45)
plt.legend(title='State')
plt.show()
```
